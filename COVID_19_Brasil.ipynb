```python
import pandas as pd
import os

df = pd.read_csv(os.path.join('data','casos-br-total.csv'), header=None)
df.columns = df.iloc[0]
df = df.iloc[1:]

df[['Suspeitos', 'Confirmados', 'Mortes']] = df[['Suspeitos', 'Confirmados', 'Mortes']].apply(pd.to_numeric)

df.head(len(df))
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Data'], df['Confirmados'].to_numpy(), linewidth=3, color='red', label=None)
plt.xticks(df['Data'], rotation=90, fontsize=12)
plt.yticks(fontsize=14)
plt.title('Número de Casos Confirmados', fontsize=21)

plt.axvline(3, 0, 8000, color='dodgerblue', linestyle='--', linewidth=2, alpha=0.7, label='Sequenciamento do Genoma')
plt.axvline(9, 0, 8000, color='k', linestyle='dotted', linewidth=2, alpha=1, label='Transmissão Comunitária')

plt.legend(loc=2, fancybox=True, fontsize=12)

plt.savefig('imagens/Confirmados-Total.png', format='png', dpi=100)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Data'], df['Mortes'].to_numpy(), linewidth=3, color='red', label=None)
plt.xticks(df['Data'], rotation=90, fontsize=12)
plt.yticks(fontsize=14)
plt.title('Número de Mortes', fontsize=21)

plt.axvline(21, 0, 8000, color='dodgerblue', linestyle='--', linewidth=2, alpha=0.7, label='Inicio das Mortes')

plt.legend(loc=2, fancybox=True, fontsize=12)

plt.savefig('imagens/Numero-Mortes-Brazil.png', format='png', dpi=100)

plt.show()
```
